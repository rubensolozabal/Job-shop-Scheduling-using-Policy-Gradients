```python
# Environment

NUM_MACHINES_DEMO = 1
NUM_JOBS_DEMO = 1

NUM_MACHINES_XSMALL = 4
NUM_JOBS_XSMALL = 4

NUM_MACHINES_SMALL = 10
NUM_JOBS_SMALL = 10

NUM_MACHINES_MEDIUM = 15
NUM_JOBS_MEDIUM = 15

NUM_MACHINES_LARGE = 20
NUM_JOBS_LARGE = 20

NUM_MACHINES_XLARGE = 25
NUM_JOBS_XLARGE = 25

MIN_TASK_TIME = 1
MAX_TASK_TIME = 10


class Config:
    
    def __init__(self):
    
        # Environment
        self.machine_profile = "small_default"
        self.job_profile = "small_default"

        self.num_jobs = 0
        self.num_tasksperjob = 0
        self.num_machines = 0  
        
        self.task_minTime = MIN_TASK_TIME
        self.task_maxTime = MAX_TASK_TIME

        self.lambda_time = 1

        self._getMachinesProperties(self.machine_profile)
        self._getJobsProperties(self.job_profile)


        # Network
        self.embedding_size = 64 # embedding size
        self.hidden_dim = 16     # agent LSTM num_neurons
        self.num_layers = 1      # agent LSTM num_stacks

        # Data
        self.batch_size = 128    # batch size


        # Training / test parameters
        self.num_epoch = 40000          # number of epochs
        self.learning_rate = 0.0005     # agent learning rate
        self.beta = 0                   # entropy regularization

        
        # Solver
        self.timeout = 1000

        
      
    def _getMachinesProperties(self, machine_profile):
       
        if machine_profile == "xsmall_default":
            
            self.num_machines = NUM_MACHINES_XSMALL
            
        elif machine_profile == "small_default":
            
            self.num_machines = NUM_MACHINES_SMALL
            
        elif machine_profile == "medium_default":

            self.num_machines = NUM_MACHINES_MEDIUM
            
        elif machine_profile == "large_default":

            self.num_machines = NUM_MACHINES_LARGE
            
        elif machine_profile == "xlarge_default":

            self.num_machines = NUM_MACHINES_XLARGE
        
        elif machine_profile == "demo_default":

            self.num_machines = NUM_MACHINES_DEMO
            
        else:
            raise Exception('Machine profile not detected.')
            
    def _getJobsProperties(self, job_profile):
       
        if job_profile == "xsmall_default":
            
            self.num_jobs = NUM_JOBS_XSMALL
            self.num_tasksperjob = NUM_MACHINES_XSMALL
            
        elif job_profile == "small_default":
            
            self.num_jobs = NUM_JOBS_SMALL
            self.num_tasksperjob = NUM_MACHINES_SMALL
            
        elif job_profile == "medium_default":

            self.num_jobs = NUM_JOBS_MEDIUM
            self.num_tasksperjob = NUM_MACHINES_MEDIUM
            
        elif job_profile == "large_default":

            self.num_jobs = NUM_JOBS_LARGE
            self.num_tasksperjob = NUM_MACHINES_LARGE
            
        elif job_profile == "xlarge_default":

            self.num_jobs = NUM_JOBS_XLARGE
            self.num_tasksperjob = NUM_MACHINES_XLARGE
        
        elif job_profile == "demo_default":

            self.num_jobs = NUM_JOBS_DEMO
            self.num_tasksperjob = NUM_MACHINES_DEMO
            
        else:
            raise Exception('Job profile not detected.')
             
            
    def reconfigure(self):
        
        self._getMachinesProperties(self.machine_profile)
        self._getJobsProperties(self.job_profile)
```
